В результате преобразования у вас должен получиться столбец, содержащий информацию о том, сколько месяцев проработал соискатель.
Выполните преобразование, ответьте на контрольные вопросы и удалите столбец **"Опыт работы"** из таблицы.


```python
def experience_to_months_simple(exp_str):
    if pd.isna(exp_str) or 'Не указано' in str(exp_str):
        return np.nan
    exp_text = str(exp_str).split(',')[0]
    years = 0
    months = 0
    words = exp_text.split()
    for i in range(len(words) - 1):
        if words[i].isdigit():
            num = int(words[i])
            next_word = words[i + 1].lower()
            
            # Проверяем следующее слово
            if any(year_word in next_word for year_word in ['лет', 'год', 'года']):
                years = num
            elif any(month_word in next_word for month_word in ['месяц', 'месяца', 'месяцев']):
                months = num
    
    return years * 12 + months
df['Опыт работы (месяц)'] = df['Опыт работы'].apply(experience_to_months_simple)
df = df.drop('Опыт работы', axis=1)
df.head()
```

Давайте создадим признаки-мигалки для каждой категории: если категория присутствует в списке желаемых соискателем, то в столбце на месте строки рассматриваемого соискателя ставится True, иначе - False.

Такой метод преобразования категориальных признаков называется One Hot Encoding и его схема представлена на рисунке ниже:
<img src=https://raw.githubusercontent.com/AndreyRysistov/DatasetsForPandas/main/ohe.jpg>

```python
employment_categories = ['полная занятость', 'частичная занятость', 'проектная работа', 'волонтерство', 'стажировка']
schedule_categories = ['полный день', 'сменный график', 'гибкий график', 'удаленная работа', 'вахтовый метод']
def create_one_hot_features(df, column_name, categories):
    for category in categories:
        new_column_name = f"{column_name}_{category.replace(' ', '_')}"
        df[new_column_name] = df[column_name].apply(
            lambda x: category in str(x) if pd.notna(x) else False
        )
    return df
df = create_one_hot_features(df, 'Занятость', employment_categories)
df = create_one_hot_features(df, 'График', schedule_categories)
employment_columns = [col for col in df.columns if col.startswith('Занятость_')]
schedule_columns = [col for col in df.columns if col.startswith('График_')]
df = df.drop(['Занятость', 'График'], axis=1)
df.head()
```

4. Умножить сумму желаемой заработной платы на присоединенный курс валюты (close) и разделить на пропорцию (обратите внимание на пропуски после объединения в этих столбцах), результат занести в новый столбец "ЗП (руб)".


```python
import pandas as pd
exchange_rates_df = pd.read_csv('data/ExchangeRates.csv')
exchange_rates_df['date'] = pd.to_datetime(exchange_rates_df['date'], format='%d/%m/%y')
exchange_rates_df = exchange_rates_df[['currency', 'date', 'close', 'proportion']]
currency_mapping = {
    'руб.': 'RUB',
    'руб': 'RUB',
    'USD': 'USD',
    'EUR': 'EUR',
    'KZT': 'KZT',
    'тенге': 'KZT',
    'KGS': 'KGS',
    'сом': 'KGS',
    'грн.': 'UAH',
    'грн': 'UAH',
    'BYN': 'BYN',
    'бел.руб': 'BYN',
    'AZN': 'AZN',
    'манат': 'AZN'
}
def extract_salary_and_currency(salary_str):
    if pd.isna(salary_str):
        return None, None
    parts = str(salary_str).split()
    if len(parts) < 2:
        return None, None
    amount_str = parts[0].replace('руб.', '').replace(' ', '')
    try:
        amount = float(amount_str)
    except:
        return None, None
    currency_str = ' '.join(parts[1:]).strip().lower()
    currency_iso = currency_mapping.get(currency_str, None)
    return amount, currency_iso
df[['salary_amount', 'salary_currency']] = df['ЗП'].apply(
    lambda x: pd.Series(extract_salary_and_currency(x))
)
df['Дата обновления'] = pd.to_datetime(df['Обновление резюме'], format='%d.%m.%Y %H:%M')
df['Дата обновления дата'] = df['Дата обновления'].dt.date
exchange_rates_df['date_date'] = exchange_rates_df['date'].dt.date
rub_rate = pd.DataFrame({
    'currency': ['RUB'],
    'date_date': [pd.to_datetime('2023-01-01').date()],
    'close': [1.0],
    'proportion': [1.0]
})
exchange_rates_df = pd.concat([exchange_rates_df, rub_rate], ignore_index=True)
exchange_rates_df = exchange_rates_df.drop_duplicates(['currency', 'date_date'])
df = pd.merge(
    df,
    exchange_rates_df,
    left_on=['salary_currency', 'Дата обновления дата'],
    right_on=['currency', 'date_date'],
    how='left'
)
df['ЗП (руб)'] = df['salary_amount'] * df['close'] / df['proportion']
df.loc[df['salary_currency'] == 'RUB', 'ЗП (руб)'] = df['salary_amount']
df = df.drop(columns=[
    'salary_amount', 'salary_currency', 'currency', 
    'date', 'close', 'proportion', 'date_date',
    'Дата обновления', 'Дата обновления дата', 'ЗП'
])
df.head()
```

6. Постройте **многоуровневую столбчатую диаграмму**, которая показывает зависимость медианной заработной платы (**"ЗП (руб)"**) от признаков **"Готовность к переезду"** и **"Готовность к командировкам"**. Проанализируйте график, сравнив уровень заработной платы в категориях.

```python
def extract_relocation(text):
    if 'готов к переезду' in text.lower() or 'готова к переезду' in text.lower():
        return 'Готов к переезду'
    elif 'не готов к переезду' in text.lower() or 'не готова к переезду' in text.lower():
        return 'Не готов к переезду'
    else:
        return 'Не указано'
def extract_business_trips(text):
    """Извлекает готовность к командировкам из текста"""
    if 'готов к командировкам' in text.lower() or 'готова к командировкам' in text.lower():
        return 'Готов к командировкам'
    elif 'готов к редким командировкам' in text.lower() or 'готова к редким командировкам' in text.lower():
        return 'Готов к редким командировкам'
    elif 'не готов к командировкам' in text.lower() or 'не готова к командировкам' in text.lower():
        return 'Не готов к командировкам'
    else:
        return 'Не указано'
df['Готовность_переезд'] = df['Город, переезд, командировки'].apply(extract_relocation)
df['Готовность_командировки'] = df['Город, переезд, командировки'].apply(extract_business_trips)
grouped = df.groupby(['Готовность_переезд', 'Готовность_командировки'])['ЗП (руб)'].median().reset_index()
pivot_table = grouped.pivot(index='Готовность_переезд', 
                           columns='Готовность_командировки', 
                           values='ЗП (руб)')
relocation_order = ['Готов к переезду', 'Не готов к переезду', 'Не указано']
business_trips_order = ['Готов к командировкам', 'Готов к редким командировкам', 
                       'Не готов к командировкам', 'Не указано']
pivot_table = pivot_table.reindex(index=relocation_order, columns=business_trips_order)
plt.figure(figsize=(12, 8))
x = np.arange(len(pivot_table.columns)) 
width = 0.25 
colors = ['#4CAF50', '#2196F3', '#FF9800']  
for i, (relocation_type, color) in enumerate(zip(pivot_table.index, colors)):
    values = pivot_table.loc[relocation_type].values
    positions = x + i * width - width  
    bars = plt.bar(positions, values, width, 
                   label=relocation_type, 
                   color=color, 
                   edgecolor='black',
                   alpha=0.8)
    for bar, value in zip(bars, values):
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2., height + 1000,
                 f'{value:,.0f}', ha='center', va='bottom', fontsize=9)
plt.xlabel('Готовность к командировкам', fontsize=12)
plt.ylabel('Медианная заработная плата (руб)', fontsize=12)
plt.title('Зависимость медианной ЗП от готовности к переезду и командировкам', 
          fontsize=14, pad=20)
plt.xticks(x, pivot_table.columns, rotation=45, ha='right')
plt.legend(title='Готовность к переезду')
plt.grid(True, alpha=0.3, axis='y')
```

```python
1. Влияние отдельных факторов:

Готовность к переезду: Готовые к переезду соискатели имеют более высокую среднюю ЗП

Готовность к командировкам: Готовые к командировкам имеют значительно более высокую среднюю ЗП

2. Наилучшие комбинации:

Наивысшая ЗП: У соискателей, готовых и к переезду, и к командировкам

Средняя ЗП: У соискателей, готовых только к командировкам

Наименьшая ЗП: У соискателей, не готовых ни к переезду, ни к командировкам

3. Взаимодействие факторов:

Оба фактора влияют на уровень ЗП

Готовность к командировкам, кажется, имеет большее влияние

Комбинация обоих факторов дает наибольший эффект
```
